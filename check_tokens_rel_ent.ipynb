```python
import json
```

```python
with open('./rel_model/assets/annotations.jsonl', 'r') as json_file:
    json_lines = list(json_file)

ent_token_start = []
rel_token_start = []
for line in json_lines:
    line = json.loads(line)
    for ent in line["spans"]:
        ent_token_start.append(ent["token_start"])
    for rel in line["relations"]:
        rel_token_start.append(rel["head_span"]["token_start"])
        rel_token_start.append(rel["child_span"]["token_start"])
    break
```

```python
print(ent_token_start)
print(rel_token_start)
```
