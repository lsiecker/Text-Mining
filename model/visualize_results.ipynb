# NER Visualization of the training results

In this notebook, we will visualize the training results of the NER model. Information is gathered about the True Positives, False Positives and False Negatives. The results are visualized in a confusion matrix and a table. Additionally the support per label is calculated and visualized.

```python
import pandas as pd 
import matplotlib.pyplot as plt

tps = pd.DataFrame({
    'TP' : [33, 4, 4, 16, 2, 8],
    'FP' : [1, 8, 14, 31, 15, 30],
    'FN' : [0, 7, 29, 39, 27, 47]},
    index = ['Datetime', 'Location', 'Cause', 'Attribute', 'Activity', 'Event'])

tps.plot(kind='bar', stacked=True, rot=0)

# labels for x & y axis
plt.ylabel('Count')
plt.xlabel('Label')
 
# title of plot
plt.title('True Positives, False Positives and False Negatives for NER - Component 2')
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

tps = pd.DataFrame({
    'TP' : [19, 1, 5, 1, 0, 0, 0, 3, 0],
    'FP' : [9, 2, 7, 0, 1, 1, 1, 3, 0],
    'FN' : [6, 4, 4, 2, 1, 7, 8, 11, 1]},
    index = ['Landmark \n Name', 'Condition', 'Date', 'Number', 'People', 'Component', 'Location', 'Type', 'Animal']) 

tps.plot(kind='bar', stacked=True, rot=0)

# labels for x & y axis
plt.ylabel('Count')
plt.xlabel('Label')
plt.xticks(rotation=45)
 
# title of plot
plt.title('True Positives, False Positives and False Negatives \n for NER - Component 1')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.cm import get_cmap
import seaborn as sns

data = {'Datetime': 33, 'Cause': 33, 'Activity': 29, 'Location': 11, 'Attribute': 55, 'Event': 55}

# Convert to DataFrame
supp = pd.DataFrame(list(data.items()), columns=['Label', 'Count'])

ax = sns.barplot(supp, x='Label', y='Count', hue='Label', legend=False)
for i in ax.containers:
    ax.bar_label(i,)
ax.set_title("Support per Label in Test Set NER - Component 2")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Original data
data = {'Positive': [67, 149], 'Negative': [99, 0]}
df_cm = pd.DataFrame(data, index=['Positive', 'Negative'])

# Mask for zeros
mask = df_cm.applymap(lambda x: x == 0)

# Replace zeros with asterisks
annot_df = df_cm.applymap(lambda x: '*' if x == 0 else f'{x}')

# Plotting with Seaborn heatmap
conf = sns.heatmap(df_cm, annot=annot_df, cmap='Blues', fmt='', cbar=True,)

# Adding labels and title
conf.set(ylabel="Predicted Values", xlabel="Actual Values", title="Confusion Matrix for NER - Component 2")

# Show the plot
plt.show()
```
