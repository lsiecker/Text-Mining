```python
import json
import os
import sys
#from preprocessor import Preprocessor 

# Load the original data from train.json
cd = os.getcwd()            # Get the current working directory
parent_d = os.path.abspath(os.path.join(cd, os.pardir))
file_name = "assets\\train.json"
file_path = os.path.join(parent_d, file_name)
with open(file_path, 'r') as f:
    data = json.load(f)
```

```python
import spacy

#TODO: gebruik ik nu een random nieuwe nlp?
def char_to_token(text, start, end):
    spacy_lib="en_core_web_sm"
    nlp = spacy.load(spacy_lib)
    doc = nlp(text)
    char_span = doc.char_span(start, end)
    if char_span is None:
        return None, None
    else:
        return char_span.start, char_span.end - 1
```

```python
# Initialize an empty list to store the converted data
converted_data = []

for item in data:
    text = item[0]

    entities_per_sentence = item[1]['entities']
    relations = item[1]['relations']

    # Entities are in format [[sentence1 label], [sentence2 label], ...] and we want them all in one list [ label1, label2, ...]
    entities = []
    for sentence in entities_per_sentence:
        entities.extend(sentence)

    # Create a dictionary for the converted data with "text", "spans", and "relations" as in github exampele	
    converted_item = {
        "text": text,
        "spans": [
            {
                "text": text, 
                "start": span[0],
                "end": span[1],
                "token_start": char_to_token(text, span[0], span[1])[0],
                "token_end": char_to_token(text, span[0], span[1])[1],
                "type": "span",
                "label": span[2],
            }
            for span in entities
        ],
        "relations": relations,
    }
    #TODO:  missing "tokens": [{"text": "Furthermore", "start": 0, "end": 11, "id": 0, "ws": false, "disabled": true}, {...
    #       missing "_input_hash": -13216227, "_task_hash": -1944660624,
    #       missing "meta": {"source": "BioNLP 2011 Genia Shared Task, PMC-1134658-00-TIAB.txt"}
    converted_data.append(converted_item)

# # Save the converted data to train_rel.json in assets folder
# folder_path = "assets"
# os.makedirs(folder_path, exist_ok=True)
# file_path = os.path.join(folder_path, 'train_rel.json')
# with open(file_path, 'w') as f:
#     json.dump(converted_data, f, indent=2)


parent_directory = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
folder_path = os.path.join(parent_directory, "assets")
os.makedirs(folder_path, exist_ok=True)
file_path = os.path.join(folder_path, 'train_rel.json')
with open(file_path, 'w') as f:
    json.dump(converted_data, f, indent=2)
```
